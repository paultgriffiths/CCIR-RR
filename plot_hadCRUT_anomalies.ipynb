```python
## import required modules
```

```python
import pylab as plt
import netCDF4 as nc
import cartopy.crs as ccrs
```

## read in data 

```python
hadcrut5 = nc.Dataset('CRUTEM.5.0.1.0.anomalies.nc')
```

## make the simplest possible plot

```python
plt.pcolormesh(hadcrut5.variables['tas'][0,:,:])
```

## try to plot the same data on a map

```python
lon = hadcrut5.variables['longitude'][:]
lat = hadcrut5.variables['latitude'][:]
data_to_plot = hadcrut5.variables['tas'][0,:,:]

ax = plt.axes(projection=ccrs.PlateCarree())
ax.set_global()
plt.pcolormesh(lon, lat, data_to_plot, transform=ccrs.PlateCarree())
ax.coastlines()
```
